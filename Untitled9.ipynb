```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Antecedents
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
food_quality = ctrl.Antecedent(np.arange(0, 11, 1), 'food_quality')

# Consequents
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')

# Fuzzy sets for the antecedents
service['poor'] = fuzz.trimf(service.universe, [0, 0, 3])
service['acceptable'] = fuzz.trimf(service.universe, [2, 5, 7])
service['amazing'] = fuzz.trimf(service.universe, [5, 10, 10])

food_quality['bad'] = fuzz.trimf(food_quality.universe, [0, 0, 3])
food_quality['decent'] = fuzz.trimf(food_quality.universe, [2, 5, 7])
food_quality['great'] = fuzz.trimf(food_quality.universe, [5, 10, 10])

# Fuzzy sets for the consequents
tip['low'] = fuzz.trimf(tip.universe, [0, 0, 8])
tip['medium'] = fuzz.trimf(tip.universe, [6, 11, 17])
tip['high'] = fuzz.trimf(tip.universe, [15, 25, 25])

# Define the fuzzy rules
rule1 = ctrl.Rule(service['amazing'] | food_quality['great'], tip['high'])
rule2 = ctrl.Rule(service['acceptable'], tip['medium'])
rule3 = ctrl.Rule(service['poor'] & food_quality['bad'], tip['low'])

# Create the control system
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])

# Create a simulation to use the control system
tipping_sim = ctrl.ControlSystemSimulation(tipping_ctrl)

# Set the input values
tipping_sim.input['service'] = 8
tipping_sim.input['food_quality'] = 6.5

# Compute the output
tipping_sim.compute()

# Get the crisp value of the tip
tip_crisp = tipping_sim.output['tip']

print("The suggested tip is:", tip_crisp, "%")
```
